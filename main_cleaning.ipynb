```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df_main = pd.read_csv("/main_df.csv")
## Data Cleaning

df_main = df_main[df_main["date"].str.contains("2013|2014|2015|2016", na=False)]
columns = [
    'date', 'lat', 'lng', 'subject_age', 'subject_race', 'subject_sex',
    'violation', 'arrest_made', 'citation_issued', 'warning_issued',
    'outcome', 'reason_for_stop', 'search_conducted', 'raw_search_warrant'
]

df_main = df_main[columns]
df_main = df_main.dropna(subset=[
    "lat",
    "lng",
    "citation_issued",
    "violation",
    "warning_issued",
    "reason_for_stop"
])

df_main['lat'] = df_main['lat'].round(decimals=6)
df_main['lng'] = df_main['lng'].round(decimals=6)
# encoding outcome column

unique_count_o = df_main['outcome'].nunique()
print(f"Number of unique values in the outcome column: {unique_count_o}")
print(df_main["outcome"].unique())
reason_map_o = {
    'warning': 1,
    'citation': 2,
    'arrest': 3
}

df_main["outcome_encd"] = df_main["outcome"].map(reason_map_o)
## Dummy Variable Encoding
# make lowercase first (important for consistency)

df_main['subject_race'] = df_main['subject_race'].str.lower()
df_main['subject_sex'] = df_main['subject_sex'].str.lower()
# one-hot encode race (force 0/1)

race_dummies = pd.get_dummies(df_main['subject_race'], prefix='race', dtype=int)
# one-hot encode sex (force 0/1)

sex_dummies = pd.get_dummies(df_main['subject_sex'], prefix='sex', dtype=int)
# add to dataframe

df_main = pd.concat([df_main, race_dummies, sex_dummies], axis=1)
# drop original encoded columns

df_main = df_main.drop(columns=['outcome'])
df_main = df_main.drop(columns=['subject_sex'])
df_main = df_main.drop(columns=['subject_race'])
# changing variable types

df_main['subject_age'] = df_main['subject_age'].astype("Int64")
df_main['outcome_encd'] = df_main['outcome_encd'].astype("Int64")
# downloading cleaned file

df_main.to_csv("/Users/caroline/Desktop/SDS 357/Data/Files/main_df_final.csv", index=False)
```
