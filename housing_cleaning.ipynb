```python
from pathlib import Path
import pandas as pd
DATA_DIR = Path("../Files")
df_housing = pd.read_csv(DATA_DIR / "housing_df.csv")
# data cleaning

df_housing = df_housing[df_housing["Sale Date"].str.contains("2013|2014|2015|2016", na=False)]
df_housing = df_housing.dropna(subset=['Property Address'])
cols = ['Property Address', 'Sale Price', 'Sale Date', 'City', 'State']
df_housing = df_housing[cols]
df_housing = df_housing.dropna()
df_housing = df_housing.rename(columns={'Property Address': 'property_address', 'Sale Price': 'sale_price', 'Sale Date': 'sale_date', 'City': 'city', 'State': 'state'})
# downloading file

df_housing.to_csv(DATA_DIR / "df_housing_clean.csv", index=False)
## after running geocoding.py complete data cleaning process 
# importing file

df_housing_clean = pd.read_csv(DATA_DIR / "df_housing_geocoded_with_zip.csv")
# removing zip code and property address columns

df_housing_clean = df_housing_clean.drop(columns=["zip_code", 'property_address'])
df_housing_clean = df_housing_clean.round({'latitude': 6})
df_housing_clean = df_housing_clean.round({'longitude': 6})
# removing any NA values 

df_housing_clean = df_housing_clean.dropna()
# renaming columns

df_housing_clean = df_housing_clean.rename(columns={'latitude': 'lat', 'longitude': 'lng'})
df_housing_clean.head()
# downloading cleaned file

df_housing_clean.to_csv(DATA_DIR / "df_housing_clean.csv", index=False)
```
